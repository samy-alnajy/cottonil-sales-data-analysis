# 📊 Cottonil HTML Reports Data Extraction Project

## Project Overview

This project focuses on extracting structured data from multiple HTML reports exported from the Cottonil system.

Each report contains tabular data distributed across many HTML pages.  
The goal of this project is to automatically parse these pages, extract the required data, clean unnecessary rows, and export the results into structured CSV files for further analysis.

The main steps performed in this notebook include:

- Reading multiple HTML files
- Parsing HTML structure using BeautifulSoup
- Extracting table data
- Cleaning unnecessary rows and headers
- Exporting the final datasets into CSV format

This automation significantly reduces manual work and prepares the data for analysis and visualization.

## 📦 Import Required Libraries

```python
from bs4 import BeautifulSoup
import pandas as pd
import os
import csv
```

## 🧾 1️⃣ Sales Extraction

```python
base_path = r"C:\Users\Ahmed\OneDrive\Documents\Data Analysis Projects\Cottonil Project\Cottonil Html Files\المبيعات"
output_file = r"C:\Users\Ahmed\OneDrive\Documents\Data Analysis Projects\Cottonil Project\Cottonil Excel Files\المبيعات\المبيعات.csv"

all_data = []

for i in range(1, 289):

    file_path = f"{base_path}\\SLSCARD08Page{i}.html"

    try:
        with open(file_path, 'r', encoding='windows-1256') as f:
            html_content = f.read()

        soup = BeautifulSoup(html_content, 'html.parser')
        rows = soup.find_all('tr')

        page_data = []

        for row in rows:
            cols = row.find_all('td')
            cols_text = [col.get_text(strip=True) for col in cols]

            if cols_text:
                page_data.append(cols_text)

        if len(page_data) > 2:
            page_data = page_data[1:-1]

        all_data.extend(page_data)

    except:
        pass

df = pd.DataFrame(all_data)
df.to_csv(output_file, index=False, encoding='utf-8-sig')

print("successful")
```

## 🛒 2️⃣ Purchases Extraction

```python
html_folder = r"C:\Users\Ahmed\OneDrive\Documents\Data Analysis Projects\Cottonil Project\Cottonil Html Files\المشتريات"
output_file = r"C:\Users\Ahmed\OneDrive\Documents\Data Analysis Projects\Cottonil Project\Cottonil Excel Files\المشتريات\المشتريات.csv"

all_rows = []

for i in range(1, 368):
    file_path = os.path.join(html_folder, f"prchcard12Page{i}.html")
    
    if not os.path.exists(file_path):
        continue

    with open(file_path, "r", encoding="windows-1256") as file:
        soup = BeautifulSoup(file, "html.parser")

    tables = soup.find_all("table")
    page_rows = []

    for table in tables:
        for tr in table.find_all("tr"):
            tds = tr.find_all("td")
            row = [td.get_text(strip=True) for td in tds]

            if len(row) >= 10:
                page_rows.append(row)

    if i > 1 and len(page_rows) > 2:
        page_rows = page_rows[2:]
    elif i > 1 and len(page_rows) <= 2:
        page_rows = []

    all_rows.extend(page_rows)

df = pd.DataFrame(all_rows)

df.to_csv(output_file, index=False, encoding="utf-8-sig")

print("successful")
```

## 📦 3️⃣ Inventory Extraction

```python
html_folder = r"C:\Users\Ahmed\OneDrive\Documents\Data Analysis Projects\Cottonil Project\Cottonil Html Files\جرد البضاعة"
output_csv = r"C:\Users\Ahmed\OneDrive\Documents\Data Analysis Projects\Cottonil Project\Cottonil Excel Files\جرد البضاعة\جرد البضاعة.csv"

all_data = []

html_files = sorted([f for f in os.listdir(html_folder) if f.endswith(".html")])

for page_index, html_file in enumerate(html_files):
    
    file_path = os.path.join(html_folder, html_file)
    
    with open(file_path, "r", encoding="windows-1256") as file:
        soup = BeautifulSoup(file, "html.parser")
    
    tables = soup.find_all("table")
    
    page_data = []
    
    for table in tables:
        rows = table.find_all("tr")
        
        for row in rows:
            cols = row.find_all(["td","th"])
            cols_text = [col.get_text(strip=True) for col in cols]
            
            if cols_text:
                page_data.append(cols_text)
    
    if page_index != 0:
        page_data = page_data[4:]
    
    if page_data:
        page_data = page_data[:-1]
    
    all_data.extend(page_data)

df = pd.DataFrame(all_data)

df.to_csv(output_csv, index=False, header=False, encoding="utf-8-sig")

print("Successful")
```

## 🔄 4️⃣ Item Movement Extraction

```python
main_folder = r"C:\Users\Ahmed\OneDrive\Documents\Data Analysis Projects\Cottonil Project\Cottonil Html Files\حركه الاصناف"
output_folder = r"C:\Users\Ahmed\OneDrive\Documents\Data Analysis Projects\Cottonil Project\Cottonil Excel Files\حركه الاصناف"

for folder_name in os.listdir(main_folder):

    folder_path = os.path.join(main_folder, folder_name)

    if os.path.isdir(folder_path):

        all_rows = []

        html_files = [f for f in os.listdir(folder_path) if f.endswith(".html")]

        for file in html_files:

            file_path = os.path.join(folder_path, file)

            with open(file_path, "r", encoding="windows-1256", errors="ignore") as f:
                soup = BeautifulSoup(f, "html.parser")

            tables = soup.find_all("table")

            for table in tables:
                rows = table.find_all("tr")

                for row in rows:
                    cols = [c.get_text(strip=True) for c in row.find_all("td")]

                    if cols:
                        all_rows.append(cols)

        df = pd.DataFrame(all_rows)

        if len(df) > 3:
            df = df.iloc[1:-2]

        output_file = os.path.join(output_folder, f"{folder_name}.csv")

        df.to_csv(output_file, index=False, encoding="utf-8-sig")

print("successful")
```

## 💰 5️⃣ Net Sales Before & After Discount

```python
input_folder = r"C:\Users\Ahmed\OneDrive\Documents\Data Analysis Projects\Cottonil Project\Cottonil Html Files\صافي المبيعات قبل وبعد الخصم"
output_file = r"C:\Users\Ahmed\OneDrive\Documents\Data Analysis Projects\Cottonil Project\Cottonil Excel Files\صافي المبيعات قبل وبعد الخصم\صافي المبيعات قبل وبعد الخصم.csv"

all_rows = []

for i in range(1, 338):

    file_path = os.path.join(input_folder, f"DAY00Page{i}.html")

    if not os.path.exists(file_path):
        continue

    with open(file_path, "r", encoding="windows-1256") as f:
        soup = BeautifulSoup(f, "html.parser")

    page_rows = []

    tables = soup.find_all("table")

    for table in tables:
        rows = table.find_all("tr")

        for row in rows:
            cols = row.find_all("font")
            data = [c.get_text(strip=True) for c in cols]

            if len(data) > 5:
                page_rows.append(data)

    if i != 1 and len(page_rows) > 0:
        page_rows = page_rows[1:]

    all_rows.extend(page_rows)

with open(output_file, "w", newline="", encoding="utf-8-sig") as f:
    writer = csv.writer(f)
    writer.writerows(all_rows)

print("successful")
```

## 📑 6️⃣ Supplier Statement Extraction

```python
folder_path = r"C:\Users\Ahmed\OneDrive\Documents\Data Analysis Projects\Cottonil Project\Cottonil Html Files\كشف حساب الموردين"
output_path = r"C:\Users\Ahmed\OneDrive\Documents\Data Analysis Projects\Cottonil Project\Cottonil Excel Files\كشف حساب الموردين\كشف حساب الموردين.csv"

all_rows = []

for i in range(1, 22):

    file_path = os.path.join(folder_path, f"spcrdfrmtorepPage{i}.html")

    with open(file_path, "r", encoding="windows-1256") as f:
        soup = BeautifulSoup(f, "html.parser")

    page_rows = []

    tables = soup.find_all("table")

    for table in tables:
        for tr in table.find_all("tr"):
            cols = [td.get_text(strip=True) for td in tr.find_all("td")]

            if len(cols) >= 9:
                page_rows.append(cols)

    if i != 1:
        page_rows = page_rows[2:]

    all_rows.extend(page_rows)

df = pd.DataFrame(all_rows)

df.to_csv(output_path, index=False, encoding="utf-8-sig")

print("successful")
```
